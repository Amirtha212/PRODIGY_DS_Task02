```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titanic_df = pd.read_csv('titanic.csv')
```

```python
print(titanic_df.head())
```

```python
print(titanic_df.isnull().sum())
```

```python
print(titanic_df.describe())
```

```python
sns.countplot(x='Survived', data=titanic_df)
plt.title('Survival Count')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Sex', data=titanic_df)
plt.title('Survival Count by Sex')
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=titanic_df)
plt.title('Survival Count by Passenger Class')
plt.show()
```

```python
sns.histplot(titanic_df['Age'].dropna(), kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.show()
```

```python
sns.histplot(x='Age', hue='Survived', data=titanic_df, kde=True)
plt.title('Survival Count by Age')
plt.xlabel('Age')
plt.show()
```

```python
sns.histplot(titanic_df['Fare'], kde=True)
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.show()
```

```python
sns.histplot(x='Fare', hue='Survived', data=titanic_df, kde=True)
plt.title('Survival Count by Fare')
plt.xlabel('Fare')
plt.show()
```

```python
numeric_df = titanic_df.select_dtypes(include=[np.number])
corr = numeric_df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
